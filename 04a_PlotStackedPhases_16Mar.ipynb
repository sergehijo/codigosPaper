### The amplitude envelope is given by magnitude of the analytic signal. 
### The instantaneous frequency can be obtained by differentiating the instantaneous phase in respect to time. 
### The instantaneous phase corresponds to the phase angle of the analytic signal.

```python
from scipy.signal import hilbert
def Hilbert_Output(signal, fs):
    analytic_signal = hilbert(signal)
    amplitude_envelope = np.abs(analytic_signal)
    #instantaneous_phase = np.unwrap(np.angle(analytic_signal), np.pi, axis=-1)
    instantaneous_phase = np.angle(analytic_signal) #% 2*np.pi
    instantaneous_frequency = (np.diff(instantaneous_phase)/(2.0*np.pi)* fs)
    inst_freq = np.lib.pad(instantaneous_frequency, (1,0), 'mean')
    return (amplitude_envelope, instantaneous_phase,inst_freq)
```

### Procedemos a realizar la transformada de Hilbert por canal 
### La idea del algoritmo es formar una matriz por canal nuevamente, que sea de ValorFaseInstantaneaxSamples
### La dimensión es la misma que la anterior de TrialsxSamples, pero sirve para hacer el StackPlot por canal después.

### Generamos el data frame que posee los valores por fase

```python
#Obtenemos el pasado
past=-0.3
#Obtenemos el futuro
future=data.tWindow.loc[0][1]
#Obtenemos la taza de sampleo
sampleRate= data.SampleRate
#Formamos el arreglo de tiempo
t=arange(past,future,1/sampleRate[1])
#Formamos el data frame a usar después
df = pd.DataFrame([], columns=['Rat',
                               'File',
                               'L-dopa',
                               'SampleRate',
                               'Bandpass',
                               'tWindow',
                               'Channel',
                               'Area',
                               'InstPhasexSamples'])

Ldopa = 'off'
rat='RAT05'
#f1 = 0.1
#f2 = 75.0
f1 = 10.0
f2 = 25.0
#For canal en canales. 
for j in range(0,len(data.Channel)):   
    #Formar una matriz de ValorFasexSamples por canal
    #Obtener las dimensiones de la matriz
    a,b = shape(data.MatrizTrialsxSamples[j])
    phaseMChannel = np.zeros([a,b])
    channel = data.MatrizTrialsxSamples[j]
    #For trial en trials. O lo que es lo mismo, por fila en la matriz de trials x Samples
    for i in range(0,a):
        #Formar una matriz de ValorFasexSamples
        #phaseMChannel = np.zeros([a,b])
        #Obtenemos el trial correspondiente
        trial = channel[i,:]
        #Obtenemos la amplitud y fase instantanea correspondiente
        amp, inst_phase, inst_freq= Hilbert_Output(trial, int(sampleRate[1]))
        phaseMChannel[i,:] = inst_phase
    #Procedemos a guardar 
    # create small df with the data of that channel
    df2 = pd.DataFrame([[rat,
                             fname,
                             Ldopa,
                             sampleRate[1],
                             [f1,f2],
                             [past,future],
                             data.Channel.loc[j],
                             data.Area.loc[j],
                             #lfparray.mean(0)]],
                             phaseMChannel]],
                       columns=['Rat',
                               'File',
                               'L-dopa',
                               'SampleRate',
                               'Bandpass',
                               'tWindow',
                               'Channel',
                               'Area',
                               'InstPhasexSamples'])
    df = pd.concat([df, df2], axis = 0)
df=df.reset_index()
del df['index']
```
